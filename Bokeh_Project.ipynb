Basic plotting using bokeh

```python
#import necesaary libraries

from bokeh.plotting import figure
from bokeh.io import output_file, show, output_notebook, push_notebook
from bokeh.models import ColumnDataSource, HoverTool, CategoricalColorMapper
from bokeh.layouts import row, column, gridplot
#from bokeh.models.widgets import Tabs, Panel
output_notebook()
```

```python
fig = figure()
fig.line([10,20,30,40,50],[60,70,80,40,50], line_width = 12)

show(fig)
```

plotting using Glyph

```python
#circle glyph

arc = figure(x_axis_label = "x axis", y_axis_label = "y axis", tools = "pan,box_zoom")
arc.circle(x=[50,40,30], y=[10,20,30], size=25, color="red", alpha=0.7, line_color="Green")

show(arc)
```

```python
#Square glyph

kit = figure(width=300, height=500)
kit.square([10,12,13,14,15],[16,17,12,14,15], size=20, color="red", alpha=0.7)

show(kit)
```

```python
#multiple lines

lines = figure(width = 400, height = 500)
lines.multi_line([[11,13,12,14],[31,23,41,13]],[[11,12,41,31],[23,45,66,78]],color=["red","green"], alpha= [0.8,0.3], line_width=5)

show(lines)
```

```python
#wedge glyph

xer = figure(width=400, height=500)
xer.wedge(x=[1,2,3], y=[1,2,3], radius=0.2, start_angle=0.4, end_angle=5.2, color="red", alpha=0.7, direction="clock" )

show(xer)
```

```python
#annular wedge glypg

rex = figure(width=400, height=400)
rex.annular_wedge(x=[10,20,30], y=[10,20,30], inner_radius=0.5, outer_radius=0.75, start_angle=0.5,
                  end_angle=5.2, color="red", alpha=0.7)

show(rex)
```

```python
#combine circle and line on a plot

x=[10,20,30,40]
y=[60,75,80,90]

ye = figure(width=400, height=500)
ye.line(x,y,line_width=2)
ye.circle(x,y,fill_color="red", size=8)

show(ye)
```

```python
#Bar plot

bar = figure(title="Bar_plot")
x = [1,2,3,4,5]
y=[1,2,3,6,7]
bar.hbar(x, right=y, height=0.6)

show(bar)
```

```python
#Vertical bar plot

v_bar = figure(title="Vertical Bar Plot")
x=[20,33,22,44]
y=[10,20,30,40]

v_bar.vbar(x, top=y, width=0.6, color="red")

show(v_bar)
```

```python
#scatter plot
import random
sca = figure(title="Scatter plot")
x = [i for i in range(234)]
y = [random.random() + 2 for i in range(234)]
sca.scatter(x,y)

show(sca)
```

```python
#area plot
import numpy as np
x=[12,20,30,40,70]
y1=[3,55,67,30,50]
y2=[1,2,3,4,5]
area = figure(width=400, height=500, title="Area Plot")
area.varea(x=x,y2=y2,y1=y1, fill_color="Red")
show(area)
```

```python
#Patch plot
x=[1,3,4,8,7]
y=[2,10,5,6,9]
Patch = figure(title="Patched Plot in bokeh", width=400, height=500)
Patch.patch(x=x,y=y, color="red")

show(Patch)
```

Map Geo data using Bokeh

```python
#necessary libraries
from bokeh.plotting import gmap
from bokeh.models import GMapOptions
```

```python
latitude1 = 40.730610
longitude1 = -73.935242
map_type="hybrid"
zoom = 14

google_map_options = GMapOptions(lat=latitude1, lng=longitude1,
                                 map_type=map_type, zoom=zoom)

title = "Newyork Map"
google_api_key = ""

google_map = gmap(google_api_key, google_map_options, title=title)

source1 = ColumnDataSource(data=dict(lat=[40.730610,40.741,40.75], lon=[-73.935242,-73.95,-73.97]))
x="lon"
y="lat"
size=30
fill_color="green"
fill_alpha=1

google_map.square(x=x,y=y,size=size,fill_color=fill_color,fill_alpha=fill_alpha,
                  source=source1)
show(google_map)
```
